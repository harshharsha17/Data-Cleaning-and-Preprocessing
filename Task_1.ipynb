checking for null values

```python
df.isnull().sum()
```

Income has 24 null values

lets fill the null values with median as it doesnt affect much with outliers

```python
median=df["Income"].median()
df["Income"]=df["Income"].fillna(value=median)
```

```python
df["Income"].isnull().sum()
```

datatypes of the column

```python
df.dtypes
```

Education,Marital_Status and Dt_Customer are objects and remaining columns are numerical columns

But Dt_Customer need to be in datetime datatype so lets change it

```python
df["Dt_Customer"]=pd.to_datetime(df["Dt_Customer"],format="%d-%m-%Y")
```

Here in Education and Marital_Status the data is not uniform so lets change it to lowercase

the data which is not uniform

```python
df[["Education","Marital_Status"]]
```

```python
df["Education"]=df["Education"].str.lower().str.strip()
df["Marital_Status"]=df["Marital_Status"].str.lower().str.strip()
```

After cleaning the data which is uniform is below

```python
df[["Education","Marital_Status"]]
```

columns of the dataset

```python
df.columns
```

columns after renaming to lowercase with no space

```python
df.columns=df.columns.str.lower().str.replace(" ","_")
df.columns
```

```python
df.info()
```

as we can see there are no null values and all columns have correct datatypes
